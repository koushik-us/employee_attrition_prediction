```python
# define list for numeric and categorical variables
numeric_columns = dataset.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = dataset.select_dtypes(include=['object', 'category']).columns

numeric_columns, categorical_columns
```

```python
#Encoding of Target variable
dataset['Attrition']=dataset['Attrition'].replace({'Yes':1,'No':0})
```

```python
dataset['Attrition'].head()
```

```python
#Encode Binary variable
dataset['OverTime']=dataset['OverTime'].map({'Yes':1,'No':0})
dataset['Gender']=dataset['Gender'].map({'Male':1,'Female':0})
```

```python
dataset['OverTime'].head()
```

```python
dataset['Gender'].head()
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
encoding_cols=['BusinessTravel','Department','EducationField','JobRole','MaritalStatus']
encoder={}
for cols in encoding_cols:
    encoder[cols]=LabelEncoder()
    dataset[cols]=encoder[cols].fit_transform(dataset[cols])
```

```python
#Correlation Matrix
corr_matrix = dataset.select_dtypes(include=['int64']).corr()
plt.figure(figsize=(20,10))
sns.heatmap(corr_matrix, cmap="crest", annot=True,)
```

```python
#Assigning Independent and Dependent variables
X = dataset.drop(['Attrition'],axis=1)
y = dataset['Attrition'].values
```

```python
y
```

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
print(Counter(y))

runs=RandomOverSampler(random_state=42)
X_over,y_over = runs.fit_resample(X,y)
print(Counter(y_over))
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X_over,y_over,test_size=0.2,random_state=42)
```

```python
# Define preprocessing steps for numerical data
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),  
    ('scaler', StandardScaler())  # Standardize numerical features
])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier,ExtraTreesClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,mean_absolute_error, classification_report
```

### Logistic Regression

```python
logClf = LogisticRegression()
pipe = Pipeline([
    ('numeric_transformer',numeric_transformer),
    ('step2',logClf)
])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('Accuracy :',accuracy_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
classification_report(y_test,y_pred)
```

### KNeighbors Classifier

```python
knClf = KNeighborsClassifier(n_neighbors=3)
pipe = Pipeline([
    ('numeric_transformer',numeric_transformer),
    ('step2',knClf)
])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('Accuracy :',accuracy_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
classification_report(y_test,y_pred)
```

### Decision Tree

```python
decisionTreeClassifier = DecisionTreeClassifier(max_depth=30)
pipe = Pipeline([
    ('numeric_transformer',numeric_transformer),
    ('step2',decisionTreeClassifier)
])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('Accuracy :',accuracy_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
classification_report(y_test,y_pred)
```

### SVM

```python
svc = SVC(kernel='rbf')
pipe = Pipeline([
    ('numeric_transformer',numeric_transformer),
    ('step2',svc)
])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('Accuracy :',accuracy_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
classification_report(y_test,y_pred)
```

### Random Forest

```python
randomForestClassifier = RandomForestClassifier(n_estimators=100,random_state=3,max_samples=0.5,max_features=0.75,max_depth=15)
pipe = Pipeline([
    ('numeric_transformer',numeric_transformer),
    ('step2',randomForestClassifier)
])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('Accuracy :',accuracy_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
classification_report(y_test,y_pred)
```

### ExtraTrees

```python
extraTreesClassifier = ExtraTreesClassifier(n_estimators=100, random_state=3, max_samples=0.5, max_features=0.75, max_depth=15, bootstrap=True)
pipe = Pipeline([
    ('numeric_transformer',numeric_transformer),
    ('step2',extraTreesClassifier)
])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('Accuracy :',accuracy_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
classification_report(y_test,y_pred)
```

### AdaBoost

```python
adaBoostClassifier = AdaBoostClassifier(n_estimators=30,learning_rate=1.0)
pipe = Pipeline([
    ('numeric_transformer',numeric_transformer),
    ('step2',adaBoostClassifier)
])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('Accuracy :',accuracy_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
classification_report(y_test,y_pred)
```

### Gradient Boost

```python
gradientBoostingClassifier= GradientBoostingClassifier(n_estimators=500)
pipe = Pipeline([
    ('numeric_transformer',numeric_transformer),
    ('step2',gradientBoostingClassifier)
])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('Accuracy :',accuracy_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
classification_report(y_test,y_pred)
```

### XGBoost

```python
xgbClassifier = XGBClassifier(n_estimators=45,max_depth=5,learning_rate=0.5)
pipe = Pipeline([
    ('numeric_transformer',numeric_transformer),
    ('step2',xgbClassifier)
])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('Accuracy :',accuracy_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
classification_report(y_test,y_pred)
```

### VotingClassifier

```python
from sklearn.ensemble import VotingClassifier, StackingClassifier

votingClassifier = VotingClassifier([('rf', randomForestClassifier), ('gbdt', gradientBoostingClassifier), ('xgb',xgbClassifier), ('et',extraTreesClassifier)],weights=[5,1,1,1])
pipe = Pipeline([
    ('numeric_transformer',numeric_transformer),
    ('step2',votingClassifier)
])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('Accuracy :',accuracy_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
classification_report(y_test,y_pred)
```

### Stacking

```python
estimators = [
    ('rf',randomForestClassifier),
    ('gbdt',gradientBoostingClassifier),
    ('xgb', xgbClassifier)
]

stackingClassifier = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression())
pipe = Pipeline([
    ('numeric_transformer',numeric_transformer),
    ('step2',stackingClassifier)
])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('Accuracy :',accuracy_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
classification_report(y_test,y_pred)
```
